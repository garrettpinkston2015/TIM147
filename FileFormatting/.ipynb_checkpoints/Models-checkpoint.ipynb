```python
import os

import pandas as pd
pd.set_option('display.max_columns', None)

import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\Garrett Pinkston\Desktop\TIM147\Project\data\RestaurantInfoComplete.csv')
df.head(3)
```

```python
df['zip'] = pd.to_numeric(df['postal_code'],errors='coerce')
df.dropna(inplace=True)
```

```python
revisedDf = df[['stars','review_count','is_open','zip','StarSlope','population','median_income','locations','chain']]
```

```python
revisedDf = pd.get_dummies(revisedDf, columns=['chain'], prefix='chain')
revisedDf.head()
```

```python
from sklearn.model_selection import train_test_split

x = revisedDf.drop('is_open',axis=1)
y = revisedDf['is_open']
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score


rf = RandomForestClassifier(n_estimators=300)
rf = rf.fit(xtrain,ytrain)
predictionRF = rf.predict(xtest)
accuracyRF = round(accuracy_score(ytest, predictionRF)*100,2)

print('Random Forest Classifier acheived max accuracy of {}% on training data using 300 estimators with a 10 K-fold Cross Validation '.format(accuracyRF))
```

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# Initialize and train the XGBoost classifier
xg_classifier = xgb.XGBClassifier(n_estimators=300)
xg_classifier.fit(xtrain, ytrain)

# Make predictions on the test set
predictions_xg = xg_classifier.predict(xtest)

# Calculate accuracy
accuracy_xg = round(accuracy_score(ytest, predictions_xg) * 100, 2)

# Print the accuracy
print('XGBoost Classifier achieved max accuracy of {}% on the test data using 300 estimators.'.format(accuracy_xg))
```

```python
#SVM
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# Assuming xtrain, xtest, ytrain, and ytest are your training and testing data
# Split the data into training and testing sets
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)

# Initialize and train the SVM classifier
svm_classifier = SVC(kernel='linear')  # You can choose different kernels like 'rbf', 'poly', etc.
svm_classifier.fit(xtrain, ytrain)

# Make predictions on the test set
predictions_svm = svm_classifier.predict(xtest)

# Calculate accuracy
accuracy_svm = round(accuracy_score(ytest, predictions_svm) * 100, 2)

# Print the accuracy
print('SVM Classifier achieved max accuracy of {}% on the test data.'.format(accuracy_svm))
```

```python
# Logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# Assuming xtrain, xtest, ytrain, and ytest are your training and testing data
# Split the data into training and testing sets
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)

# Initialize and train the Logistic Regression classifier
logreg_classifier = LogisticRegression()
logreg_classifier.fit(xtrain, ytrain)

# Make predictions on the test set
predictions_logreg = logreg_classifier.predict(xtest)

# Calculate accuracy
accuracy_logreg = round(accuracy_score(ytest, predictions_logreg) * 100, 2)

# Print the accuracy
print('Logistic Regression Classifier achieved max accuracy of {}% on the test data.'.format(accuracy_logreg))
```
